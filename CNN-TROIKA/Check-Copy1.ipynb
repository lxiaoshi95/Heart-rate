```python
dataset = sio.loadmat('testA_1.mat')
Y_test = dataset['Y_test']
X_testA = dataset['X_testA']
dataset = sio.loadmat('testB_1.mat')
Y_test = dataset['Y_test']
X_testB = dataset['X_testB']
model = ori_model(0.25)
model.load_weights("person1.hdf5")
```

```python
raw_predict, new_predict, raw = post_process(model, X_testA, X_testB, Y_test)
Y_error = abs(new_predict - Y_test)

plt.figure
plt.plot(Y_test)
plt.plot(raw)
plt.show()
Y_error = abs(raw - Y_test)
print('The avarage predict error: ', np.mean(Y_error))
print('The mean squared error:', metrics.mean_squared_error(Y_test, raw))
plt.figure
plt.plot(Y_test)
plt.plot(raw_predict)
plt.show()
Y_error = abs(raw_predict - Y_test)
print('The avarage predict error: ', np.mean(Y_error))
print('The mean squared error:', metrics.mean_squared_error(Y_test, raw_predict))
plt.figure
plt.plot(Y_test)
plt.plot(new_predict)
plt.show()
Y_error = abs(new_predict - Y_test)
print('The avarage predict error: ', np.mean(Y_error))
print('The mean squared error:', metrics.mean_squared_error(Y_test, new_predict))

sio.savemat('person_1.mat', {'gt': Y_test, 'hr': new_predict})
```
